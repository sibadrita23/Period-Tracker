```python
def calculate_next_period(last_period_date, cycle_length):
    next_period_date = last_period_date + timedelta(days=cycle_length)
    fertile_start_date = last_period_date + timedelta(days=cycle_length - 14)
    fertile_end_date = fertile_start_date + timedelta(days=6)
    return next_period_date, fertile_start_date, fertile_end_date
```
